```python
# 1. import libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# 2. find url and store it in a variable
url = "https://www.acclaimedmusic.net/070731/1948-02s.htm"

# 3. download html with a get request
response = requests.get(url)
response.status_code # 200 status code means OK!

# 4.1. parse html (create the 'soup')
soup = BeautifulSoup(response.content, "html.parser")
# 4.2. check that the html code looks like it should
soup
```

```python
artists = []

for i in range(0, 800, 4):
    
    artist_element = soup.select("tr td a")[i]
    
    if artist_element:
        artist = artist_element.get_text().strip()
        artists.append(artist)

print(artists)
```

```python
titles = []

for i in range(0, 800, 4):
    
    title_element = soup.select("tr td a")[i+1]
    
    if title_element:
        title = title_element.get_text().strip()
        titles.append(title)

print(titles)
```

```python
links = []

for tag in soup.select("div a"):
    links.append('https://www.acclaimedmusic.net/070731/' + tag["href"])

links
```

```python
titles = []

for url_name in links:
    url = url_name
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    

    
    for i in range(0, 800, 4):
    
        artist_element = soup.select("tr td a")[i]
        
        if artist_element:
            artist = artist_element.get_text().strip()
            artists.append(artist)

print(artists)
```

```python
titles = []

for url_name in links:
    url = url_name
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    

    for i in range(0, 800, 4):
    
        title_element = soup.select("tr td a")[i+1]
    
        if title_element:
            title = title_element.get_text().strip()
            titles.append(title)

print(titles)  
```

```python
df = pd.DataFrame({'artist': artists, 'title': titles})
df
```

```python
df.to_csv('not_hot.csv')
```
